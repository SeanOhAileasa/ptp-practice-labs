###### ``nslookup``
***

``nslookup`` is a program to query the Internet Domain Name System (DNS) via submitting requests to domain name servers. 

``nslookup`` operates in one of two modes:

> ``Interactive mode``

- allows a user to search domain name servers for IP and name information on various hosts or domains. 


- can also be used in this mode to display a list of hosts in a domain.

> ``Non-interactive mode``

- just returns the name or IP information requested for an individual host or domain.

> ``Terminal``

If you simply type in the name of most Linux utilities at the prompt without any options, it will return a screen of basic help and use information. But this does not work for ``nslookup`` since just entering ``nslookup`` will start the program in interactive mode. 

To get help and use information for ``nslookup`` you have to visit the ``man`` (short for manual) page.

> ``man nslookup``

The most basic way to use ``nslookup`` is to run it with the name of a host or domain you want to look up - displaying the IP address associated with ``www.comptia.org``:

> ``nslookup www.comptia.org``

IP address associated with ``comptia.org``:

- ``104.18.17.29``

Uses the default domain name server your computer is set up to use. 

To use a different domain name server, simply add a hyphen (``-``) and the server name or IP address.

Using Google's public domain name server at ``8.8.8.8``.

> ``nslookup www.comptia.org - 8.8.8.8``

Both the default and Google domain name servers provided the same IP address for ``comptia.org``.

The queries may well return different results. In some circumstances, this can indicate an issue with DNS (perhaps a record update has not been fully propagated and some resolvers have an old record cached). In this case, if you analyze the domains associated with serving the site, you will notice the use of ``trafficmanager.net``. This site uses a content delivery network (CDN) for load balancing.

Can also perform a reverse DNS look up by providing an IP address.

> ``nslookup 104.18.17.29``

In this instance, no reverse lookup record is configured, so the query returns ``NXDOMAN`` (non-existent domain). 

To launch ``nslookup`` in interactive mode, you simply need to run the ``nslookup`` command.

In interactive mode, you are presented with a greater than sign ( ``>`` ) prompt - can run simple searches and stay in interactive mode.

> ``nslookup``

Type ``www.comptia.org`` and press the ENTER key. As before, this might return a different IP address to the previous result (``104.18.17.29``).

The power of interactive mode comes in via being able to provide options and search for different types of DNS records. For example, you can arrange to return only ``MX`` (mail exchange) records with the ``set type=mx`` option.

> ``set type=mx``

This returns to the prompt - now when a search is conducted it will only return ``MX`` records.

> ``comptia.org``

The fully qualified domain name of the mail exchanger host is: 


- ``comptia-org.mail.protection.outlook.com``

Setting the query type back to returning host records.

> ``set type=a``

Now running a query to return the IPv4 address of the ``MX`` record:

- ``comptia-org.mail.protection.outlook.com``

Notice that the host record resolves to more than one IP address - this is another load balancing mechanism.

The IPv4 address of the ``MX`` record:

- ``104.47.56.110``

Now running a reverse lookup for the IP address of the mail host.

A reverse lookup record is configured - ``PTR`` records are used by other hosts to verify the sender and are very important to configure correctly for mail services.

To exit interactive mode: 

> ``exit``

###### ``dig``
***

``Domain Information Groper``, is another program used for querying the Domain Name System (DNS) via domain name servers. 

``dig`` allows a user to search for information on different domain record types, including hosts address, mail exchange records, and DNS servers associated with a domain. 

It is commonly used for DNS troubleshooting and for domain reconnaissance.

> ``Terminal``

As with ``nslookup``, you can not get help and use information from just typing ``dig`` at the prompt, as this will return a list of root DNS servers in use. 

To get use and help information you most visit the manual page.

> ``man dig``

The most basic way to use ``dig`` is to run it with the name of a host or domain you want to look up - displaying the IP address associated with ``comptia.org``:

> ``dig comptia.org``

The power of ``dig`` comes in when using its options. For example, we can find other DNS record types by using the ``-t`` option - will search for ``NS`` (name server) records.

> ``dig ns comptia.org``

Reviewing the ``man`` page, ``dig`` has a number of additional options such as being able to query a particular DNS using ``@servername`` and the ``-f`` option which lets you read in a list of queries to be done in batch mode.

###### ``whois``
***

``whois`` is another powerful tool for gathering information about a domain and related Internet information.


- The ``whois`` system provides information on domain names, IP address blocks, and autonomous systems.


- It is a query and response protocol and can be accessed via command line tools and many websites including most domain registrars.


While many domain registrars sell or provide private or anonymous registration services, there is still much useful information which can be found out using ``whois``.

> ``Terminal``

``whois`` is not by default installed on [ [Windows](https://docs.microsoft.com/en-us/sysinternals/downloads/whois) ].

Like with most Linux utilities (and unlike ``nslookup`` and ``dig``) you can get help and use information from just typing ``whois`` at the prompt.

> ``whois``

The most basic way to use ``whois`` is to run it with the name of a domain you want to look up - displaying the ``whois`` information associated with ``comptia.org``:

> ``whois comptia.org``

There is a lot of information returned on the domain, so you will have to scroll back up to see the top of the output.

- ``whois`` records contain information about the creation and expiration dates of domains and sometimes the contact information of the person or entity that registered the domain.


- The contact information is less useful these days as many domain registrars sell private registration services which hides this information.

The ``comptia.org`` domain was first registered in 1995.

``whois`` records also contain information about:

- location where the domain was registered


- one of the most important data points which can't be hidden, the DNS servers used by the domain.

``whois`` can be a useful tool to start with when doing recon on a potential penetration target. 

Getting the DNS servers allows one to use the other tools (``nslookup`` and ``dig``), to find out additional information on the target such as their web and email servers, etc.

###### Summary
***

> ``nslookup`` is used for searching DNS records.

- Modes: 

i. Interactive


ii. Non-interactive

> ``dig`` is used for searching DNS records.

- ``-t`` option in ``dig`` will select the DNS record type to return.

> ``whois`` is used for searching Domain Registration Records.

- ``whois`` can provide the creation and expiration dates and registrant contact information and list the DNS servers of the domain.
