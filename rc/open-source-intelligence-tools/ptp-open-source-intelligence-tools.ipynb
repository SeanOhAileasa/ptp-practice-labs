###### ``Maltego``
***

``Maltego`` is an information gathering and data analysis tool - can discover all kinds of information on a target and visualize how it is all connected together. 

``Maltego`` requires a user to create an account and full Internet access so can not be fully utilized in this lab - see a message about Memory optimization, click ``Restart later``.

###### ``recon-ng``
***

``Recon-ng`` is a reconnaissance framework for gathering information from web based sources. 

It looks and feels similar to ``Metasploit``. It has a module based approach and initially comes with no modules loaded as seen by clicking on ``recon-ng`` to launch it.

> ``help``

Note, the choice for the ``marketplace`` - could install all the modules with the command ``marketplace install all``.

###### ``Google Dorks``
***

> ``Firefox ESR`` > ``www.google.com``

> Search ``site:comptia.org ext:pdf`` to see if ``comptia.org`` has any PDFs:

> ``site:comptia.org ext:pdf``

Can use the ``ext:`` option with just about any type of file such as documents (``.doc`` or ``.docx``), spreadsheets (``.xls`` or ``.xlsx``) or a ``filetype`` option - can be useful during reconnaissance. 

> ``site:comptia.org filetype:pdf``

Useful Google dorking options include ``inurl:`` and ``intitle:`` which can look for any keyword in the URL or Title of a page on a website - see if ``comptia.org`` has any login pages:

> ``site:comptia.org inurl:login``

Not surprisely, find that ``comptia.org`` has a number of login pages - should be able to see the value of these options when preparing for a penetration test.
